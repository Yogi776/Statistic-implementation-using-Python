# Percentile

## What is use of Percentile in Data Science?

Suppose I am working E-Commerce Comapany,

Taking delivery more than 5 days:- 
    
    In delivery 1 day, 2 days, 3 days, 4 days, 5 days .................. n days.
    
    if I Find 5 days to we used percentile.
    
    

```python
print('25 percent value is : ',np.percentile(data_male['Weight'],25))
print('50 percent value is : ',np.percentile(data_male['Weight'],50))
print('75 percent value is : ',np.percentile(data_male['Weight'],75))
print('90 percent value is : ',np.percentile(data_male['Weight'],90))
print('100 percent value is : ',np.percentile(data_male['Weight'],100))
```

# Cumulative Density Function

## 1. What is pdf?

### Ans:-  Smoother version of histrogram

#### Where Y axis is Probability Value

## Example:- 

```python
#getting data from histrogram
count,bins_count = np.histogram(data_male['Weight'],bins=10)

#Finding pdf of the Histrogram using Count values
pdf = count / sum(count)

# using numpy np.cusum to calculate cdf 
cdf = np.cumsum(pdf)

#Ploting pdf and cdf

plt.plot(bins_count[1:],pdf,color = 'red',label = 'PDF')
plt.plot(bins_count[1:],cdf,label = 'CDF')
plt.show()
```

```python
#getting data from histrogram
count,bins_count = np.histogram(data_male['Height'],bins=10)

#Finding pdf of the Histrogram using Count values
pdf = count / sum(count)

# using numpy np.cusum to calculate cdf 
cdf = np.cumsum(pdf)

#Ploting pdf and cdf

plt.plot(bins_count[1:],pdf,color = 'red',label = 'PDF')
plt.plot(bins_count[1:],cdf,label = 'CDF')
plt.show()
```
