## How to convert Standard Deviation

```python
print("Mean of sepal_length: ",data['sepal_length'])
print("Mean of sepal_width: ",data['sepal_width'])
print("Mean of petal_length: ",data['petal_length'])
print("Mean of petal_width: ",data['petal_width'])
```

```python
print(np.std(data['sepal_length']))
print(np.std(data['sepal_width']))
print(np.std(data['petal_length']))
print(np.std(data['petal_width']))
```

```python
data['sepal_length'] = np.std(data['sepal_length'])
```

```python
data['sepal_width'] = np.std(data['sepal_width'])
data['petal_length'] = np.std(data['petal_length'])
data['petal_width'] = np.std(data['petal_width'])
```
