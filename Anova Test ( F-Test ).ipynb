```python
df_anova = df[['petal_width','species']]
```

```python
 group = df['species'].unique()
```

```python
group
```

```python
d_data = {group:df_anova['petal_width'][df_anova.species == group] for group in group}
```

```python
import scipy.stats as stats
```

```python
d_data
```

```python
F,p = stats.f_oneway(d_data['setosa'],d_data['versicolor'],d_data['versicolor'])
```

```python
F
```

```python
p
```

```python
if p< 0.05:
    print("Rejected Null Hypothesis")
else:
    print("Accepted Null Hyposthesis")
```
